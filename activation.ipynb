```python
import torch
import matplotlib.pyplot as plt

# Define a range of input values
x = torch.linspace(-5, 5, 100, dtype=torch.float)

# Activation functions and their titles
activation_functions = [
    (torch.nn.ReLU(), "ReLU"),
    (torch.nn.Sigmoid(), "Sigmoid"),
    (torch.nn.Tanh(), "Tanh"),
    (torch.nn.LeakyReLU(negative_slope=0.3), "Leaky ReLU α=0.3")
]

# Create a 2x2 grid of subplots
fig, axs = plt.subplots(2, 2, figsize=(10, 8))

# Flatten the array of axes for easy iterating
axs_flat = axs.flatten()

for ax, (activation, title) in zip(axs_flat, activation_functions):
    y = activation(x)
    ax.plot(x.numpy(), y.numpy())
    ax.set_title(title)
    ax.set_xlabel("Dane wejściowe")
    ax.set_ylabel("Dane wyjściowe")
    ax.grid(True)

# Adjust layout to not overlap plots
plt.tight_layout()
plt.show()
```
